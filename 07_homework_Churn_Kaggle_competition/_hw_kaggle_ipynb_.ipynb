# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score, roc_curve

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.compose import ColumnTransformer
```

```python
# Нормировка данных
scaler = StandardScaler()

X_train, X_valid, y_train, y_valid = train_test_split(data[feature_cols],data[target_col],train_size=0.8, random_state=42)

train_dummies = pd.get_dummies(X_train[feature_cols],columns=cat_cols)
train_preprocessed = scaler.fit_transform(train_dummies)

valid_dummies = pd.get_dummies(X_valid[feature_cols],columns=cat_cols)
valid_preprocessed = scaler.transform(valid_dummies)

test_dummies = pd.get_dummies(test,columns=cat_cols)
test_preprocessed = scaler.transform(test_dummies)
```

```python
cls = LogisticRegression()
params = {'C': [100.0, 10.0, 1.0, 0.1, 0.01, 0.001]}
grid_cls = GridSearchCV(cls,
                        param_grid=params,
                        scoring='roc_auc',
                        cv=5,
                        n_jobs= -1,
                        verbose=10,
                        )

grid_cls.fit(train_preprocessed,y_train)

cls_best = grid_cls.best_estimator_
cls_best.fit(train_preprocessed,y_train)
```

```python
y_train_predicted = cls_best.predict_proba(train_preprocessed)[:, 1]
y_valid_predicted = cls_best.predict_proba(valid_preprocessed)[:, 1]

train_auc = roc_auc_score(y_train, y_train_predicted)
valid_auc = roc_auc_score(y_valid, y_valid_predicted)

plt.figure(figsize=(10,7))
plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_valid, y_valid_predicted)[:2], label='valid AUC={:.4f}'.format(valid_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```

```python
import catboost
```

```python
X_train,X_valid,y_train,y_valid = train_test_split(data[feature_cols],data[target_col],train_size=0.8, random_state=42)
```

```python
boosting_model = catboost.CatBoostClassifier(min_data_in_leaf=1, max_depth=3, learning_rate=0.08, 
                                l2_leaf_reg=3.75, random_state=42, iterations=300,
                                loss_function='Logloss',
                                eval_metric='AUC',verbose = 0)

boosting_model.fit(X_train,y_train,cat_features = cat_cols)
```

```python
y_train_predicted = boosting_model.predict_proba(X_train)[:, 1]
y_valid_predicted = boosting_model.predict_proba(X_valid)[:, 1]

train_auc = roc_auc_score(y_train, y_train_predicted)
valid_auc = roc_auc_score(y_valid, y_valid_predicted)

plt.figure(figsize=(10,7))
plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_valid, y_valid_predicted)[:2], label='valid AUC={:.4f}'.format(valid_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```
